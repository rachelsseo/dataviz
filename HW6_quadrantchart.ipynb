```python
# data 
df = pd.read_csv('../code-python/data/nfelo-rankings.csv')
df.head(25)
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import urllib.request
from PIL import Image
import ssl


# create an SSL context that does not verify certificates - had trouble opening images
# from the web
context = ssl._create_unverified_context()

# change font family
plt.rcParams['font.family'] = 'Georgia'

# plot 
plt.figure(figsize=(12, 8))
plt.figure(dpi=170) 
plt.scatter(df['epa_per_play_against_ytd'], df['epa_per_play_ytd'], color='#ffffff', alpha=0)

# title and labels
plt.title('No One Should Be Surprised The Eagles Won The Super Bowl', fontsize=12)
plt.ylabel('Offensive EPA/play', fontsize=6)
plt.xlabel('Defensive EPA/play', fontsize=6)

# logos
for i in range(len(df)):
    if df['team_nick'][i] == 'Jets':
        continue # bypassing the Jets logo
    url = df['team_logo_espn'][i]
    with urllib.request.urlopen(url, context=context) as response:
        logo = np.array(Image.open(response))
    imagebox = OffsetImage(logo, zoom=0.05)
    ab = AnnotationBbox(imagebox, (df['epa_per_play_against_ytd'][i], df['epa_per_play_ytd'][i]), frameon=False)
    plt.gca().add_artist(ab)

# adding Jets logo
url = df['team_logo_espn'][df['team_nick'] == 'Jets'].values[0] # converts into numpy array to access first element
with urllib.request.urlopen(url, context=context) as response:
    logo = np.array(Image.open(response))
imagebox = OffsetImage(logo, zoom=0.01)
ab = AnnotationBbox(imagebox, (df['epa_per_play_against_ytd'][df['team_nick'] == 'Jets'].values[0], df['epa_per_play_ytd'][df['team_nick'] == 'Jets'].values[0]), frameon=False)
plt.gca().add_artist(ab)

# axis limits
plt.xlim(.25, -.25)
plt.ylim(-.25, .25)
plt.xticks([])
plt.yticks([])

# quadrant lines
plt.axvline(0, color='gray', linestyle='--')
plt.axhline(0, color='gray', linestyle='--')

# removing all spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

# adding arrows along x axis
plt.annotate('', xy=(-0.25, -0.25), xytext=(-0.04, -0.25), arrowprops=dict(arrowstyle='->', lw=1.5, color='green'))
plt.annotate('', xy=(0.25, -0.25), xytext=(0.04, -0.25), arrowprops=dict(arrowstyle='->', lw=1.5, color='red'))

# adding arrows along y axis
plt.annotate('', xy=(0.25, -0.25), xytext=(0.25, -0.05), arrowprops=dict(arrowstyle='->', lw=1.5, color='red'))
plt.annotate('', xy=(0.25, 0.25), xytext=(0.25, 0.05), arrowprops=dict(arrowstyle='->', lw=1.5, color='green'))

# create circle around the Eagles logo
eagles_x = df['epa_per_play_against_ytd'][df['team_nick'] == 'Eagles'].values[0]
eagles_y = df['epa_per_play_ytd'][df['team_nick'] == 'Eagles'].values[0]
circle = plt.Circle((eagles_x, eagles_y), 0.02, color='green', fill=False, lw=2)
plt.gca().add_artist(circle)

# annotation for Eagles
plt.figtext(0.7, 0.72, 'The Eagles land in the top right quadrant,', ha='left', fontsize=5)
plt.figtext(0.7, 0.705, 'meaning they lead in both offense and defense.', ha='left', fontsize=5)

# create circle around Chiefs logo
chiefs_x = df['epa_per_play_against_ytd'][df['team_nick'] == 'Chiefs'].values[0]
chiefs_y = df['epa_per_play_ytd'][df['team_nick'] == 'Chiefs'].values[0]
circle = plt.Circle((chiefs_x, chiefs_y), 0.02, color='red', fill=False, lw=2)
plt.gca().add_artist(circle)

# source
plt.figtext(0.1, -0.02, 'Source: https://www.nfeloapp.com/nfl-power-ratings/', ha='left', fontsize=5)

plt.show()
```
